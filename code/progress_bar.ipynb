## Track progress of Stark units computed

```python
import pandas as pd
import json
```

```python
res = pd.read_csv('../data/res.txt', sep=':', index_col=0)
res_exp = pd.read_csv('../data/res_exp.txt', sep=':', index_col=0)
#data = pd.read_csv('data.csv')
#data = data[['Label', 'Projective image']]
```

### Overall progress

```python
total = len(res)
found = total - (res['Stark unit, b=1'] == '-').sum()
tried = (res_exp['L-value'] != '-').sum()
mod_problem = (res_exp['Stark unit, b=1'] == 'Mod problem').sum()
error_problem = (res_exp['Stark unit, b=1'] == 'error').sum()

print('Total Stark units found: ', found, '/', total)
print('Percentage: ', found/total*100, '%')
print('')
print('Success rate without mod:', found, '/', tried)
print('Percentage: ', found/tried*100, '%')
print('')
print('Mod problem: ', mod_problem, '/', (tried+mod_problem))
print('Percentage: ', mod_problem/(tried+mod_problem)*100, '%')
print('')
print('Unidentified errors:', error_problem, '/', tried+error_problem)
print('Percentage: ', error_problem/(tried+error_problem)*100, '%')
```

### Sorted by degree of coefficient field

```python
# Stats
first = 19306
stats = pd.DataFrame(columns=['Size Galois', 'found', 'tried', 'total', 'total SR', 'tried SR', 'mod problem', 'error'])
res_exp = pd.read_csv('res_exp.txt', sep=':', index_col=0)
total, found, tried = 0, 0, 0
for i in range(2, 6):
    total_p, found_p, mod_p, error_p, tried_p = 0, 0, 0, 0, 0
    for index, row in res_exp.loc[:first].iterrows():
        if len(json.loads(row['coef. polynomial'])) == i:
            total_p = total_p + 1
            if row['Stark unit, b=1'] not in ['-', 'Mod problem', 'error']:
                found_p = found_p + 1
            if row['Stark unit, b=1'] == 'Mod problem':
                mod_p = mod_p + 1
            if row['Stark unit, b=1'] == 'error':
                error_p = error_p + 1
            if row['L-value'] != '-' or row['Stark unit, b=1'] in ['Mod problem', 'error']:
                tried_p = tried_p + 1
    if total_p > 0:
        sr_total_p = float(found_p/total_p*100)
    else:
        sr_total_p = 0
    if tried_p > 0:
        sr_tried_p = float(found_p/tried_p*100)
    else:
        sr_tried_p = 0
    
    new_row = pd.DataFrame({'Size Galois': [i-1], 'found': [found_p], 'tried': [tried_p], 'total': [total_p], 'total SR': [sr_total_p], 'tried SR': [sr_tried_p], 'mod problem': [mod_p], 'error': [error_p]})
    stats = pd.concat([stats, new_row], ignore_index=True)
    found = found + found_p
    tried = tried + tried_p
print('Success:', found, '/', first, "Percentage:", float(found/first*100), '%')
print('Success for tried:', found, '/', tried, "Percentage:", float(found/tried*100), '%')
display(stats)
```
